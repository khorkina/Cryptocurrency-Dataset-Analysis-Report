```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

data=pd.read_csv(r"C:\Users\newyo\Downloads\Crypto.csv\Crypto.csv")
```

```python
# Distribution of Cryptocurrency Prices
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], kde=True, bins=50)
plt.title('Distribution of Cryptocurrency Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Convert marketcap to numeric, removing any commas
data['marketcap'] = data['marketcap'].str.replace(',', '').astype(float)

# Top 10 Cryptocurrencies by Market Cap
top10_marketcap = data.nlargest(10, 'marketcap')
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='marketcap', data=top10_marketcap)
plt.title('Top 10 Cryptocurrencies by Market Cap')
plt.xlabel('Cryptocurrency')
plt.ylabel('Market Cap')
plt.xticks(rotation=45)
plt.show()
```

```python
# 24-Hour Trading Volume vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='volume24hrs', data=data)
plt.title('24-Hour Trading Volume vs Price')
plt.xlabel('Price')
plt.ylabel('24-Hour Trading Volume')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
# Market Cap Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['marketcap'], kde=True, bins=50)
plt.title('Market Cap Distribution')
plt.xlabel('Market Cap')
plt.ylabel('Frequency')
plt.xscale('log')
plt.show()
```

```python
# Top 10 Cryptocurrencies by Circulating Supply
top10_circulating = data.nlargest(10, 'circulatingsupply')
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='circulatingsupply', data=top10_circulating)
plt.title('Top 10 Cryptocurrencies by Circulating Supply')
plt.xlabel('Cryptocurrency')
plt.ylabel('Circulating Supply')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
corr = data[['price', 'volume24hrs', 'marketcap', 'circulatingsupply', 'maxsupply', 'totalsupply']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Price vs Circulating Supply
plt.figure(figsize=(10, 6))
sns.scatterplot(x='circulatingsupply', y='price', data=data)
plt.title('Price vs Circulating Supply')
plt.xlabel('Circulating Supply')
plt.ylabel('Price')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
# Price Time Series for Top Cryptocurrency
top_crypto = data.nlargest(1, 'marketcap')['name'].values[0]
top_crypto_data = data[data['name'] == top_crypto]

plt.figure(figsize=(10, 6))
plt.plot(pd.to_datetime(top_crypto_data['date_taken']), top_crypto_data['price'])
plt.title(f'Price Time Series for {top_crypto}')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Max Supply Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['maxsupply'], kde=True, bins=50)
plt.title('Max Supply Distribution')
plt.xlabel('Max Supply')
plt.ylabel('Frequency')
plt.xscale('log')
plt.show()
```

```python
# Total Supply vs Market Cap
plt.figure(figsize=(10, 6))
sns.scatterplot(x='totalsupply', y='marketcap', data=data)
plt.title('Total Supply vs Market Cap')
plt.xlabel('Total Supply')
plt.ylabel('Market Cap')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
